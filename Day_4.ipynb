## 3. For Loop with if Condition Inside

```python
# Even or Odd using for loop
for num in range(1, 11):
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
l = [2, 45, 78, 12, 12, 12, 12, 'Himanshu', "khokhar" , (6+7j), [56, 67, 78, 'dsfdsf']]
```

```python
for i in l:
    if type(i) == str:
        l1 = []
        for j in i:
            l1.append(j)
        print(l1)
```

##  4. While Loop – Basic Example

```python
# While loop from 1 to 5
count = 1
while count <= 5:
    print("Count is:", count)
    count += 1
```

```python
a = 1
while a<6:
    print(a)
    a = a+1
```

```python
a = 1
while a<6:
    print(a)
    if a == 4:
        break
    a = a+1
```
